#### ( 2 ): Count the number of hyperlink URLs (which begin with http) in the <u>head</u> & <u>body</u> tags.

##### [20 points]

```python
# Enter your code here, printing relevant answers to console:
head_links = htmlsoup.head.find_all('a', href=True)
http_links_in_head = [link for link in head_links if link['href'].startswith('http')]
print("Number of http links in <head>: ", len(http_links_in_head))
```

```python
body_links = htmlsoup.body.find_all('a', href=True)
http_links_in_body = [link for link in body_links if link['href'].startswith('http')]
print("Number of http links in <body>: ", len(http_links_in_body))
```

```python
import pandas as pd
```

```python
# Enter your code here, printing relevant answers to console:
http_links_in_body = set(link.get('href') for link in htmlsoup.body.find_all('a', href=lambda x: x and x.startswith('http')))
url_series = pd.Series(list(http_links_in_body))
url_series
```

```python
from urllib.parse import urlparse

def extract_domain(url):
    parsed_url = urlparse(url)
    domain = parsed_url.netloc
    return domain

extract_domain("https://www.washingtonpost.com/world/2023/10/29/israel-ground-war-gaza/")
```

```python
domain_series = url_series.map(extract_domain)
domain_series
```
