#### ( 1a ) Fetch the online data from the following location and convert it into a python JSON object:
- https://data.cityofnewyork.us/api/views/25th-nujf/rows.json?accessType=DOWNLOAD
- please use the packages below which we covered in class (urllib and json)

####  (1b) Create a new variable of type ***python list*** called "name_data" which are the values associated with the JSON key = 'data'.


```python
# Enter your code here, printing relevant answers to console:
import urllib
import pandas as pd
import json
from pprint import pprint
```

```python
url = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.json?accessType=DOWNLOAD'
response = urllib.request.urlopen(url)
response.code
json_string = response.read().decode('utf-8')
parsed_json = json.loads(json_string)
parsed_json
```

```python
name_data = parsed_json.get("data", [])
name_data
```

#### ( 2a ): Create a pandas dataframe from the values in "name_data" list

    - year (2011)

```python
# Enter your code here, printing relevant answers to console:
columns_to_extract = list(range(8, 13))
filtered_data = [[row[i] for i in columns_to_extract] for row in name_data]

df = pd.DataFrame(filtered_data)
df = df.rename(columns = {0: 'year',1:'gender',2:'ethnicity',3:'name',4:'instance of the name'})

numeric_columns = ['year', 'instance of the name']
df[numeric_columns] = df[numeric_columns].astype(float)
df
```

#### ( 3a ):  Create a new dataframe by aggregating the dataframe from (2a) to the following level:
- Group by Gender, Name and return the sum of name instances (name[12] from 2a)

#### ( 3b ): Count the number of distinct names in the dataframe created in (3a).
- If Alex is a FEMALE and a MALE name, still only count this name once.

#### ( 3c ): Create a list of distinct names which are assigned to both FEMALE and MALE children and print the length of the list.


```python
# Enter your code here, printing relevant answers to console:
grouped_df = df.groupby(['gender', 'name'])['instance of the name'].sum().reset_index()
grouped_df['name'].nunique()
```

```python
both_genders_names = grouped_df[grouped_df['gender'].isin(['FEMALE', 'MALE'])]
both_genders_names = both_genders_names.groupby('name').filter(lambda x: len(x) == 2)['name'].unique()
both_genders_names
len(both_genders_names)
```
